```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', style='size', size='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker', size='size')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', style='size', size='smoker', col='time')
```

```python
sns.scatterplot(x='day', y='total_bill', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.catplot(x='day', y='tip', data=tips)
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='swarm', hue='size')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='box', hue="sex")
```
